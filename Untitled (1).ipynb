```python
import pandas as pd

file_path = "KG_R_datay.xlsx"   # already uploaded
df = pd.read_excel(file_path)

# Normalize column names
df.columns = df.columns.str.strip().str.lower()

print(df.columns)
df.head()
```

```python
edges = []
df_subset = df.sample(n=10000, random_state=42)

for _, row in df_subset.iterrows():
    
    title = str(row['title'])
    year = str(row['year'])
    source = str(row['source'])
    publisher = str(row['publisher'])
    place = str(row['place'])
    url = str(row['articleurl'])
    
    # Handle multiple authors
    authors = str(row['authors']).split(',')
    
    for author in authors:
        author = author.strip()
        
        # Author → Title
        edges.append((author, title, "wrote"))
    
    # Title → Source
    edges.append((title, source, "published_in"))
    
    # Source → Publisher
    edges.append((source, publisher, "published_by"))
    
    # Publisher → Place
    edges.append((publisher, place, "located_in"))
    
    # Title → Year
    edges.append((title, year, "year"))
    
    # Title → URL
    edges.append((title, url, "has_url"))
```

```python
import networkx as nx

G = nx.DiGraph()

for src, tgt, rel in edges:
    G.add_edge(src, tgt, relationship=rel)

print("Nodes:", len(G.nodes()))
print("Edges:", len(G.edges()))
```

```python
from pyvis.network import Network

net = Network(
    notebook=True,
    height="600px",
    width="100%",
    cdn_resources='in_line'
)

for node in G.nodes:
    net.add_node(str(node), label=str(node))

for u, v, data in G.edges(data=True):
    net.add_edge(str(u), str(v), label=str(data['relationship']))

# ✅ FIX: manual write with UTF-8
html = net.generate_html()

with open("knowledge_graph.html", "w", encoding="utf-8") as f:
    f.write(html)
```

```python
def search_graph(query):
    results = []
    
    for u, v, data in G.edges(data=True):
        if query.lower() in u.lower() or query.lower() in v.lower():
            results.append((u, data['relationship'], v))
    
    return results
```

```python
def search_by_relation(relation):
    return [(u, data['relationship'], v) 
            for u, v, data in G.edges(data=True) 
            if relation.lower() in data['relationship'].lower()]
```

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('all-MiniLM-L6-v2')

edge_texts = [f"{u} {data['relationship']} {v}" for u, v, data in G.edges(data=True)]
edge_embeddings = model.encode(edge_texts, convert_to_tensor=True)
```

```python
def semantic_search(query, top_k=5):
    query_emb = model.encode(query, convert_to_tensor=True)
    scores = util.cos_sim(query_emb, edge_embeddings)[0]
    
    top_results = scores.argsort(descending=True)[:top_k]
    
    return [edge_texts[i] for i in top_results]
```

```python
while True:
    q = input("Enter query (or 'exit'): ")
    if q == "exit":
        break
    
    results = search_graph(q)
    
    for r in results[:10]:
        print(r)
```
