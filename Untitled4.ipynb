```python
"""
Knowledge Graph Construction from Academic Publications Dataset
===============================================================
This script builds a comprehensive knowledge graph from academic publication data,
enabling search, visualization, and analysis of research relationships.
"""

import pandas as pd
import numpy as np
import networkx as nx
import streamlit as st
from pyvis.network import Network
import plotly.graph_objects as go
import plotly.express as px
from collections import Counter
import re
from itertools import combinations
import warnings
warnings.filterwarnings('ignore')

# ============================================================================
# PART 1: DATA LOADING AND PREPROCESSING
# ============================================================================

class DataPreprocessor:
    """Handle data loading and preprocessing operations."""

    def __init__(self):
        self.df = None
        self.cleaned_df = None

    def load_data(self, file_path=None, uploaded_file=None):
        """Load data from file path or uploaded file."""
        if uploaded_file is not None:
            self.df = pd.read_excel(uploaded_file)
        elif file_path is not None:
            self.df = pd.read_excel(file_path)
        else:
            raise ValueError("Either file_path or uploaded_file must be provided")

        print(f"Loaded {len(self.df)} records")
        return self.df

    def preprocess_data(self):
        """Clean and preprocess the dataset."""
        self.cleaned_df = self.df.copy()

        # Handle missing values
        self.cleaned_df['Year'] = pd.to_numeric(self.cleaned_df['Year'], errors='coerce')
        self.cleaned_df['Cites'] = pd.to_numeric(self.cleaned_df['Cites'], errors='coerce').fillna(0)

        # Extract authors as list
        self.cleaned_df['Authors_List'] = self.cleaned_df['Authors'].apply(
            lambda x: [a.strip() for a in str(x).split(';')] if pd.notna(x) else []
        )

        # Extract keywords as list
        self.cleaned_df['Keywords_List'] = self.cleaned_df['Keywords'].apply(
            lambda x: [k.strip().lower() for k in str(x).split(';')] if pd.notna(x) else []
        )

        # Extract countries from Publisher/Place
        country_patterns = {
            'India': ['India', 'Bangalore', 'New Delhi', 'Hyderabad'],
            'USA': ['USA', 'United States', 'America', 'Chicago', 'Lincoln'],
            'UK': ['United Kingdom', 'UK', 'England', 'London'],
            'Germany': ['Germany', 'Berlin Heidelberg'],
            'Canada': ['Canada'],
            'Sweden': ['Sweden'],
            'South Korea': ['South Korea', 'Korea'],
            'Netherlands': ['Netherlands'],
        }

        def extract_country(place, publisher):
            text = str(place) + ' ' + str(publisher)
            for country, patterns in country_patterns.items():
                for pattern in patterns:
                    if pattern in text:
                        return country
            return 'Other'

        self.cleaned_df['Country'] = self.cleaned_df.apply(
            lambda x: extract_country(x.get('Place', ''), x.get('Publisher', '')), axis=1
        )

        # Extract source type
        self.cleaned_df['Source_Type'] = self.cleaned_df['Source'].apply(
            lambda x: 'Journal' if 'Journal' in str(x) else
                     'Conference' if 'Conference' in str(x) or 'Proceedings' in str(x) else 'Other'
        )

        print(f"Preprocessed {len(self.cleaned_df)} records")
        return self.cleaned_df

    def get_statistics(self):
        """Generate basic statistics about the dataset."""
        stats = {
            'total_publications': len(self.cleaned_df),
            'total_citations': int(self.cleaned_df['Cites'].sum()),
            'avg_citations': round(self.cleaned_df['Cites'].mean(), 2),
            'unique_authors': len(set([a for authors in self.cleaned_df['Authors_List'] for a in authors if a])),
            'unique_keywords': len(set([k for keywords in self.cleaned_df['Keywords_List'] for k in keywords if k])),
            'year_range': f"{int(self.cleaned_df['Year'].min())} - {int(self.cleaned_df['Year'].max())}",
            'journals': self.cleaned_df['Source'].nunique(),
            'countries': self.cleaned_df['Country'].nunique()
        }
        return stats


# ============================================================================
# PART 2: KNOWLEDGE GRAPH CONSTRUCTION
# ============================================================================

class KnowledgeGraphBuilder:
    """Build knowledge graph from preprocessed data."""

    def __init__(self, df):
        self.df = df
        self.graph = nx.MultiDiGraph()

    def build_graph(self):
        """Build the complete knowledge graph."""
        print("Building knowledge graph...")

        # Add paper nodes
        for idx, row in self.df.iterrows():
            paper_id = f"P{idx}"
            self.graph.add_node(
                paper_id,
                type='paper',
                title=row['Title'],
                year=int(row['Year']) if pd.notna(row['Year']) else None,
                citations=int(row['Cites']) if pd.notna(row['Cites']) else 0,
                source=row['Source'],
                publisher=row['Publisher'],
                country=row['Country'],
                abstract=str(row['Abstract'])[:500] if pd.notna(row['Abstract']) else ''
            )

        # Add author nodes and paper-author edges
        author_nodes = {}
        for idx, row in self.df.iterrows():
            paper_id = f"P{idx}"
            for author in row['Authors_List']:
                if author and author != 'nan':
                    if author not in author_nodes:
                        author_id = f"A{len(author_nodes)}"
                        author_nodes[author] = author_id
                        self.graph.add_node(author_id, type='author', name=author)
                    self.graph.add_edge(author_nodes[author], paper_id, type='wrote')

        # Add keyword nodes and paper-keyword edges
        keyword_nodes = {}
        for idx, row in self.df.iterrows():
            paper_id = f"P{idx}"
            for keyword in row['Keywords_List']:
                if keyword and keyword != 'nan':
                    if keyword not in keyword_nodes:
                        keyword_id = f"K{len(keyword_nodes)}"
                        keyword_nodes[keyword] = keyword_id
                        self.graph.add_node(keyword_id, type='keyword', name=keyword)
                    self.graph.add_edge(paper_id, keyword_nodes[keyword], type='has_keyword')

        # Add citation edges (paper cites paper)
        # Using citation counts as a proxy for citation relationships
        # For more accurate citations, we would need the actual citation network data

        # Add co-authorship edges
        for idx, row in self.df.iterrows():
            authors = row['Authors_List']
            if len(authors) >= 2:
                for author1, author2 in combinations(authors, 2):
                    if author1 in author_nodes and author2 in author_nodes:
                        self.graph.add_edge(
                            author_nodes[author1],
                            author_nodes[author2],
                            type='co_author',
                            weight=1
                        )

        # Add journal nodes and paper-journal edges
        journal_nodes = {}
        for idx, row in self.df.iterrows():
            paper_id = f"P{idx}"
            journal = row['Source']
            if pd.notna(journal) and journal != 'nan':
                if journal not in journal_nodes:
                    journal_id = f"J{len(journal_nodes)}"
                    journal_nodes[journal] = journal_id
                    self.graph.add_node(journal_id, type='journal', name=journal)
                self.graph.add_edge(paper_id, journal_nodes[journal], type='published_in')

        # Add country nodes and paper-country edges
        country_nodes = {}
        for idx, row in self.df.iterrows():
            paper_id = f"P{idx}"
            country = row['Country']
            if country and country != 'Other':
                if country not in country_nodes:
                    country_id = f"C{len(country_nodes)}"
                    country_nodes[country] = country_id
                    self.graph.add_node(country_id, type='country', name=country)
                self.graph.add_edge(paper_id, country_nodes[country], type='from_country')

        print(f"Graph built with {self.graph.number_of_nodes()} nodes and {self.graph.number_of_edges()} edges")
        return self.graph

    def get_node_types(self):
        """Return count of each node type."""
        node_types = {}
        for node, data in self.graph.nodes(data=True):
            ntype = data.get('type', 'unknown')
            node_types[ntype] = node_types.get(ntype, 0) + 1
        return node_types


# ============================================================================
# PART 3: GRAPH ENRICHMENT AND ANALYSIS
# ============================================================================

class GraphAnalyzer:
    """Perform analysis on the knowledge graph."""

    def __init__(self, graph, df):
        self.graph = graph
        self.df = df

    def calculate_metrics(self):
        """Calculate various graph metrics."""
        metrics = {}

        # Basic metrics
        metrics['nodes'] = self.graph.number_of_nodes()
        metrics['edges'] = self.graph.number_of_edges()
        metrics['density'] = nx.density(self.graph)

        # Degree metrics
        degrees = dict(self.graph.degree())
        metrics['avg_degree'] = np.mean(list(degrees.values())) if degrees else 0
        metrics['max_degree'] = max(degrees.values()) if degrees else 0

        # Find top nodes by degree
        top_nodes = sorted(degrees.items(), key=lambda x: x[1], reverse=True)[:10]
        metrics['top_nodes'] = [(self.graph.nodes[n].get('name', n), d) for n, d in top_nodes]

        # Connected components
        components = list(nx.connected_components(self.graph.to_undirected()))
        metrics['num_components'] = len(components)
        metrics['largest_component_size'] = len(max(components, key=len)) if components else 0

        return metrics

    def find_top_authors(self, n=10):
        """Find top authors by number of publications."""
        author_pubs = {}
        for node, data in self.graph.nodes(data=True):
            if data.get('type') == 'author':
                # Count edges to paper nodes
                pubs = sum(1 for _, _, edge_data in self.graph.edges(node, data=True)
                          if edge_data.get('type') == 'wrote')
                author_pubs[data.get('name', node)] = pubs

        return sorted(author_pubs.items(), key=lambda x: x[1], reverse=True)[:n]

    def find_top_keywords(self, n=15):
        """Find most frequent keywords."""
        keyword_counts = {}
        for node, data in self.graph.nodes(data=True):
            if data.get('type') == 'keyword':
                keyword_counts[data.get('name', node)] = self.graph.degree(node)

        return sorted(keyword_counts.items(), key=lambda x: x[1], reverse=True)[:n]

    def find_top_countries(self, n=10):
        """Find countries with most publications."""
        country_counts = {}
        for node, data in self.graph.nodes(data=True):
            if data.get('type') == 'country':
                country_counts[data.get('name', node)] = self.graph.degree(node)

        return sorted(country_counts.items(), key=lambda x: x[1], reverse=True)[:n]

    def find_top_journals(self, n=10):
        """Find most productive journals."""
        journal_counts = {}
        for node, data in self.graph.nodes(data=True):
            if data.get('type') == 'journal':
                journal_counts[data.get('name', node)] = self.graph.degree(node)

        return sorted(journal_counts.items(), key=lambda x: x[1], reverse=True)[:n]

    def find_collaboration_network(self):
        """Find the largest collaboration component."""
        # Build author collaboration graph
        collab_graph = nx.Graph()

        for u, v, data in self.graph.edges(data=True):
            if data.get('type') == 'co_author':
                collab_graph.add_edge(u, v)

        # Find largest connected component
        if collab_graph.number_of_nodes() > 0:
            components = list(nx.connected_components(collab_graph))
            largest = max(components, key=len)
            return largest
        return set()

    def search_papers(self, query, search_type='title'):
        """Search for papers matching the query."""
        query = query.lower()
        results = []

        for idx, row in self.df.iterrows():
            paper_id = f"P{idx}"
            match = False

            if search_type == 'title' and query in str(row['Title']).lower():
                match = True
            elif search_type == 'author':
                for author in row['Authors_List']:
                    if query in author.lower():
                        match = True
                        break
            elif search_type == 'keyword':
                for keyword in row['Keywords_List']:
                    if query in keyword.lower():
                        match = True
                        break
            elif search_type == 'abstract' and query in str(row['Abstract']).lower():
                match = True
            else:  # all fields
                if (query in str(row['Title']).lower() or
                    query in str(row['Abstract']).lower() or
                    any(query in a.lower() for a in row['Authors_List']) or
                    any(query in k.lower() for k in row['Keywords_List'])):
                    match = True

            if match:
                results.append({
                    'paper_id': paper_id,
                    'title': row['Title'],
                    'authors': ', '.join(row['Authors_List'][:3]) + ('...' if len(row['Authors_List']) > 3 else ''),
                    'year': row['Year'],
                    'citations': row['Cites'],
                    'source': row['Source'],
                    'keywords': ', '.join(row['Keywords_List'][:5])
                })

        return results


# ============================================================================
# PART 4: VISUALIZATION MODULE
# ============================================================================

class GraphVisualizer:
    """Create visualizations of the knowledge graph."""

    def __init__(self, graph, df):
        self.graph = graph
        self.df = df

    def create_interactive_network(self, max_nodes=100, node_types=None):
        """Create an interactive network visualization using PyVis."""
        net = Network(height="750px", width="100%", bgcolor="#ffffff", font_color="black")

        # Filter nodes
        nodes_to_include = set()
        if node_types:
            for node, data in self.graph.nodes(data=True):
                if data.get('type') in node_types:
                    nodes_to_include.add(node)
        else:
            # Get highest degree nodes
            degrees = dict(self.graph.degree())
            nodes_to_include = set(sorted(degrees, key=degrees.get, reverse=True)[:max_nodes])

        # Add nodes
        for node in nodes_to_include:
            data = self.graph.nodes[node]
            node_type = data.get('type', 'unknown')

            # Node color mapping
            colors = {
                'paper': '#3498db',      # blue
                'author': '#e74c3c',     # red
                'keyword': '#2ecc71',    # green
                'journal': '#f39c12',     # orange
                'country': '#9b59b6',    # purple
                'unknown': '#95a5a6'     # gray
            }

            # Node size based on degree
            degree = self.graph.degree(node)
            size = 15 + min(degree * 2, 50)

            # Node title for tooltip
            title = f"{data.get('name', node)}\nType: {node_type}\nConnections: {degree}"
            if node_type == 'paper':
                title += f"\nCitations: {data.get('citations', 0)}\nYear: {data.get('year', 'N/A')}"

            net.add_node(
                node,
                label=data.get('name', node)[:20],
                title=title,
                color=colors.get(node_type, colors['unknown']),
                size=size,
                shape='dot'
            )

        # Add edges
        for u, v, data in self.graph.edges(data=True):
            if u in nodes_to_include and v in nodes_to_include:
                edge_type = data.get('type', 'related')
                edge_color = {
                    'wrote': '#e74c3c',
                    'has_keyword': '#2ecc71',
                    'published_in': '#f39c12',
                    'co_author': '#3498db',
                    'from_country': '#9b59b6'
                }.get(edge_type, '#95a5a6')

                net.add_edge(u, v, color=edge_color, title=edge_type)

        # Set physics layout
        net.set_options("""
        var options = {
            nodes: {
                font: { size: 12 }
            },
            edges: {
                smooth: { enabled: true }
            },
            physics: {
                stabilization: { iterations: 100 },
                barnesHut: { gravitationalConstant: -8000 }
            }
        }
        """)

        return net

    def create_publication_timeline(self):
        """Create a timeline of publications over years."""
        year_counts = self.df['Year'].value_counts().sort_index()

        fig = go.Figure(data=[
            go.Bar(
                x=year_counts.index.astype(int),
                y=year_counts.values,
                marker_color='#3498db',
                hovertemplate='Year: %{x}<br>Publications: %{y}<extra></extra>'
            )
        ])

        fig.update_layout(
            title='Publications Over Time',
            xaxis_title='Year',
            yaxis_title='Number of Publications',
            template='plotly_white',
            height=400
        )

        return fig

    def create_top_authors_chart(self, top_authors):
        """Create bar chart for top authors."""
        names, counts = zip(*top_authors) if top_authors else ([], [])

        fig = go.Figure(data=[
            go.Bar(
                x=list(names)[:10],
                y=list(counts)[:10],
                marker_color='#e74c3c',
                orientation='v',
                hovertemplate='Author: %{x}<br>Publications: %{y}<extra></extra>'
            )
        ])

        fig.update_layout(
            title='Top Authors by Publication Count',
            xaxis_title='Author',
            yaxis_title='Number of Publications',
            template='plotly_white',
            height=500,
            xaxis_tickangle=-45
        )

        return fig

    def create_top_keywords_chart(self, top_keywords):
        """Create horizontal bar chart for top keywords."""
        keywords, counts = zip(*top_keywords) if top_keywords else ([], [])

        fig = go.Figure(data=[
            go.Bar(
                x=list(counts)[:15],
                y=list(keywords)[:15],
                marker_color='#2ecc71',
                orientation='h',
                hovertemplate='Keyword: %{y}<br>Occurrences: %{x}<extra></extra>'
            )
        ])

        fig.update_layout(
            title='Top Keywords by Frequency',
            xaxis_title='Number of Occurrences',
            yaxis_title='Keyword',
            template='plotly_white',
            height=500
        )

        return fig

    def create_country_distribution(self, top_countries):
        """Create pie chart for country distribution."""
        countries, counts = zip(*top_countries) if top_countries else ([], [])

        fig = go.Figure(data=[
            go.Pie(
                labels=list(countries),
                values=list(counts),
                hole=0.3,
                marker=dict(colors=px.colors.qualitative.Set3),
                hovertemplate='%{label}<br>Publications: %{value}<br>Percentage: %{percent}<extra></extra>'
            )
        ])

        fig.update_layout(
            title='Publication Distribution by Country',
            template='plotly_white',
            height=450
        )

        return fig

    def create_citation_distribution(self):
        """Create histogram of citation distribution."""
        citations = self.df[self.df['Cites'] > 0]['Cites']

        fig = go.Figure(data=[
            go.Histogram(
                x=citations,
                nbinsx=30,
                marker_color='#f39c12',
                hovertemplate='Citations: %{x}<br>Count: %{y}<extra></extra>'
            )
        ])

        fig.update_layout(
            title='Citation Distribution',
            xaxis_title='Number of Citations',
            yaxis_title='Number of Papers',
            template='plotly_white',
            height=400
        )

        return fig

    def create_network_summary(self):
        """Create a summary dashboard of network metrics."""
        # Count node types
        node_types = {}
        for node, data in self.graph.nodes(data=True):
            ntype = data.get('type', 'unknown')
            node_types[ntype] = node_types.get(ntype, 0) + 1

        fig = go.Figure(data=[
            go.Bar(
                x=list(node_types.keys()),
                y=list(node_types.values()),
                marker_color=['#3498db', '#e74c3c', '#2ecc71', '#f39c12', '#9b59b6'],
                hovertemplate='Type: %{x}<br>Count: %{y}<extra></extra>'
            )
        ])

        fig.update_layout(
            title='Network Node Distribution by Type',
            xaxis_title='Node Type',
            yaxis_title='Number of Nodes',
            template='plotly_white',
            height=400
        )

        return fig


# ============================================================================
# PART 5: STREAMLIT APPLICATION
# ============================================================================

class KnowledgeGraphApp:
    """Main Streamlit application for Knowledge Graph Explorer."""

    def __init__(self):
        self.preprocessor = None
        self.graph = None
        self.analyzer = None
        self.visualizer = None
        self.df = None

    def run(self):
        st.set_page_config(page_title="Knowledge Graph Explorer", layout="wide", page_icon="📚")

        st.title("📚 Academic Knowledge Graph Explorer")
        st.markdown("---")

        # Sidebar
        with st.sidebar:
            st.header("📁 Data Upload")
            uploaded_file = st.file_uploader(
                "Upload Excel/CSV File",
                type=['xlsx', 'xls', 'csv'],
                help="Upload your academic publications dataset"
            )

            st.markdown("---")
            st.header("🔍 Search Options")
            search_type = st.selectbox(
                "Search Field",
                ['title', 'author', 'keyword', 'abstract', 'all fields']
            )
            search_query = st.text_input("Enter search query", placeholder="e.g., AI, blockchain, library")

            if st.button("🔎 Search", type="primary"):
                if search_query and self.analyzer:
                    self.display_search_results(search_query, search_type)

            st.markdown("---")
            st.header("🎨 Visualization Settings")
            max_nodes = st.slider("Max Nodes in Network", 50, 500, 150)
            node_types_filter = st.multiselect(
                "Show Node Types",
                ['paper', 'author', 'keyword', 'journal', 'country'],
                default=['paper', 'author', 'keyword']
            )

            if st.button("🔄 Refresh Graph", use_container_width=True):
                if self.graph:
                    st.session_state['refresh_graph'] = True

        # Main content area
        col1, col2 = st.columns([2, 1])

        with col1:
            st.subheader("📊 Knowledge Graph Overview")

            if uploaded_file is not None:
                if 'graph_built' not in st.session_state or st.session_state.get('refresh_graph'):
                    with st.spinner("Loading and processing data..."):
                        self.build_knowledge_graph(uploaded_file)
                    st.session_state['graph_built'] = True
                    st.session_state['refresh_graph'] = False

                if self.graph:
                    # Display statistics
                    stats = self.preprocessor.get_statistics() if self.preprocessor else {}

                    metric_cols = st.columns(4)
                    metrics_data = [
                        ("📄 Publications", stats.get('total_publications', 0)),
                        ("✏️ Authors", stats.get('unique_authors', 0)),
                        ("🏷️ Keywords", stats.get('unique_keywords', 0)),
                        ("📅 Years", stats.get('year_range', 'N/A'))
                    ]

                    for col, (label, value) in zip(metric_cols, metrics_data):
                        with col:
                            st.metric(label, value)

                    # Interactive network visualization
                    st.markdown("### 🌐 Interactive Knowledge Graph")
                    net = self.visualizer.create_interactive_network(
                        max_nodes=max_nodes,
                        node_types=node_types_filter if node_types_filter else None
                    )
                    net.save_graph("temp_network.html")
                    with open("temp_network.html", "r", encoding="utf-8") as f:
                        st.components.v1.html(f.read(), height=600)

                    # Analytics section
                    st.markdown("### 📈 Analytics Dashboard")
                    tab1, tab2, tab3, tab4 = st.tabs(
                        ["📅 Timeline", "🏆 Top Authors", "🔑 Keywords", "🌍 Countries"]
                    )

                    with tab1:
                        fig = self.visualizer.create_publication_timeline()
                        st.plotly_chart(fig, use_container_width=True)

                    with tab2:
                        top_authors = self.analyzer.find_top_authors(15)
                        fig = self.visualizer.create_top_authors_chart(top_authors)
                        st.plotly_chart(fig, use_container_width=True)
                        with st.expander("View full author list"):
                            for author, count in top_authors:
                                st.write(f"**{author}**: {count} publications")

                    with tab3:
                        top_keywords = self.analyzer.find_top_keywords(20)
                        fig = self.visualizer.create_top_keywords_chart(top_keywords)
                        st.plotly_chart(fig, use_container_width=True)

                    with tab4:
                        top_countries = self.analyzer.find_top_countries()
                        fig = self.visualizer.create_country_distribution(top_countries)
                        st.plotly_chart(fig, use_container_width=True)

            else:
                st.info("👈 Please upload your dataset file to begin")
                st.markdown("""
                ### Expected Data Format
                The dataset should contain columns such as:
                - **Title**: Paper title
                - **Authors**: Semicolon-separated author names
                - **Year**: Publication year
                - **Cites**: Number of citations
                - **Keywords**: Semicolon-separated keywords
                - **Source**: Journal/Conference name
                - **Publisher**: Publishing organization
                - **Abstract**: Paper abstract
                """)

        with col2:
            st.subheader("📋 Graph Statistics")

            if self.graph:
                metrics = self.analyzer.calculate_metrics()

                st.markdown("#### Network Metrics")
                st.metric("Total Nodes", metrics['nodes'])
                st.metric("Total Edges", metrics['edges'])
                st.metric("Network Density", f"{metrics['density']:.4f}")
                st.metric("Average Degree", f"{metrics['avg_degree']:.2f}")

                st.markdown("#### Node Type Distribution")
                node_types = self.visualizer.create_network_summary()
                st.plotly_chart(node_types, use_container_width=True)

                st.markdown("#### Top Connected Nodes")
                for node_name, degree in metrics['top_nodes'][:5]:
                    st.write(f"• {node_name}: {degree} connections")

                st.markdown("#### Recent Publications")
                recent_pubs = self.df.nlargest(5, 'Year') if self.df is not None else []
                for _, row in recent_pubs.iterrows():
                    with st.expander(f"📄 {row['Title'][:50]}..."):
                        st.write(f"**Authors**: {row['Authors'][:100]}...")
                        st.write(f"**Year**: {row['Year']}")
                        st.write(f"**Citations**: {row['Cites']}")
                        st.write(f"**Source**: {row['Source']}")
            else:
                st.info("Upload data to view statistics")

    def build_knowledge_graph(self, uploaded_file):
        """Build the knowledge graph from uploaded file."""
        self.preprocessor = DataPreprocessor()
        self.preprocessor.load_data(uploaded_file=uploaded_file)
        self.df = self.preprocessor.preprocess_data()

        builder = KnowledgeGraphBuilder(self.preprocessor.cleaned_df)
        self.graph = builder.build_graph()

        self.analyzer = GraphAnalyzer(self.graph, self.preprocessor.cleaned_df)
        self.visualizer = GraphVisualizer(self.graph, self.preprocessor.cleaned_df)

    def display_search_results(self, query, search_type):
        """Display search results."""
        st.markdown("---")
        st.subheader(f"🔍 Search Results: '{query}'")

        results = self.analyzer.search_papers(query, search_type)

        if results:
            st.success(f"Found {len(results)} matching papers")

            for i, result in enumerate(results[:20]):  # Limit to 20 results
                with st.expander(f"📄 {result['title']}"):
                    col1, col2 = st.columns([3, 1])
                    with col1:
                        st.write(f"**Authors**: {result['authors']}")
                        st.write(f"**Source**: {result['source']}")
                        st.write(f"**Keywords**: {result['keywords']}")
                    with col2:
                        st.write(f"**Year**: {result['year']}")
                        st.write(f"**Citations**: {result['citations']}")

            if len(results) > 20:
                st.info(f"Showing first 20 of {len(results)} results")
        else:
            st.warning("No results found. Try a different search query.")


# ============================================================================
# MAIN EXECUTION
# ============================================================================

def main():
    app = KnowledgeGraphApp()
    app.run()


if __name__ == "__main__":
    main()
```

```python
%%writefile app.py
"""
Knowledge Graph Construction from Academic Publications Dataset
===============================================================
This script builds a comprehensive knowledge graph from academic publication data,
enabling search, visualization, and analysis of research relationships.
"""

import pandas as pd
import numpy as np
import networkx as nx
import streamlit as st
from pyvis.network import Network
import plotly.graph_objects as go
import plotly.express as px
from collections import Counter
import re
from itertools import combinations
import warnings
warnings.filterwarnings('ignore')

# ============================================================================
# PART 1: DATA LOADING AND PREPROCESSING
# ============================================================================

class DataPreprocessor:
    """Handle data loading and preprocessing operations."""

    def __init__(self):
        self.df = None
        self.cleaned_df = None

    def load_data(self, file_path=None, uploaded_file=None):
        """Load data from file path or uploaded file."""
        if uploaded_file is not None:
            self.df = pd.read_excel(uploaded_file)
        elif file_path is not None:
            self.df = pd.read_excel(file_path)
        else:
            raise ValueError("Either file_path or uploaded_file must be provided")

        print(f"Loaded {len(self.df)} records")
        return self.df

    def preprocess_data(self):
        """Clean and preprocess the dataset."""
        self.cleaned_df = self.df.copy()

        # Handle missing values
        self.cleaned_df['Year'] = pd.to_numeric(self.cleaned_df['Year'], errors='coerce')
        self.cleaned_df['Cites'] = pd.to_numeric(self.cleaned_df['Cites'], errors='coerce').fillna(0)

        # Extract authors as list
        self.cleaned_df['Authors_List'] = self.cleaned_df['Authors'].apply(
            lambda x: [a.strip() for a in str(x).split(';')] if pd.notna(x) else []
        )

        # Extract keywords as list
        self.cleaned_df['Keywords_List'] = self.cleaned_df['Keywords'].apply(
            lambda x: [k.strip().lower() for k in str(x).split(';')] if pd.notna(x) else []
        )

        # Extract countries from Publisher/Place
        country_patterns = {
            'India': ['India', 'Bangalore', 'New Delhi', 'Hyderabad'],
            'USA': ['USA', 'United States', 'America', 'Chicago', 'Lincoln'],
            'UK': ['United Kingdom', 'UK', 'England', 'London'],
            'Germany': ['Germany', 'Berlin Heidelberg'],
            'Canada': ['Canada'],
            'Sweden': ['Sweden'],
            'South Korea': ['South Korea', 'Korea'],
            'Netherlands': ['Netherlands']
        }

        def extract_country(place, publisher):
            text = str(place) + ' ' + str(publisher)
            for country, patterns in country_patterns.items():
                for pattern in patterns:
                    if pattern in text:
                        return country
            return 'Other'

        self.cleaned_df['Country'] = self.cleaned_df.apply(
            lambda x: extract_country(x.get('Place', ''), x.get('Publisher', '')), axis=1
        )

        # Extract source type
        self.cleaned_df['Source_Type'] = self.cleaned_df['Source'].apply(
            lambda x: 'Journal' if 'Journal' in str(x) else
                     'Conference' if 'Conference' in str(x) or 'Proceedings' in str(x) else 'Other'
        )

        print(f"Preprocessed {len(self.cleaned_df)} records")
        return self.cleaned_df

    def get_statistics(self):
        """Generate basic statistics about the dataset."""
        stats = {
            'total_publications': len(self.cleaned_df),
            'total_citations': int(self.cleaned_df['Cites'].sum()),
            'avg_citations': round(self.cleaned_df['Cites'].mean(), 2),
            'unique_authors': len(set([a for authors in self.cleaned_df['Authors_List'] for a in authors if a])),
            'unique_keywords': len(set([k for keywords in self.cleaned_df['Keywords_List'] for k in keywords if k])),
            'year_range': f"{int(self.cleaned_df['Year'].min())} - {int(self.cleaned_df['Year'].max())}",
            'journals': self.cleaned_df['Source'].nunique(),
            'countries': self.cleaned_df['Country'].nunique()
        }
        return stats


# ============================================================================
# PART 2: KNOWLEDGE GRAPH CONSTRUCTION
# ============================================================================

class KnowledgeGraphBuilder:
    """Build knowledge graph from preprocessed data."""

    def __init__(self, df):
        self.df = df
        self.graph = nx.MultiDiGraph()

    def build_graph(self):
        """Build the complete knowledge graph."""
        print("Building knowledge graph...")

        # Add paper nodes
        for idx, row in self.df.iterrows():
            paper_id = f"P{idx}"
            self.graph.add_node(
                paper_id,
                type='paper',
                title=row['Title'],
                year=int(row['Year']) if pd.notna(row['Year']) else None,
                citations=int(row['Cites']) if pd.notna(row['Cites']) else 0,
                source=row['Source'],
                publisher=row['Publisher'],
                country=row['Country'],
                abstract=str(row['Abstract'])[:500] if pd.notna(row['Abstract']) else ''
            )

        # Add author nodes and paper-author edges
        author_nodes = {}
        for idx, row in self.df.iterrows():
            paper_id = f"P{idx}"
            for author in row['Authors_List']:
                if author and author != 'nan':
                    if author not in author_nodes:
                        author_id = f"A{len(author_nodes)}"
                        author_nodes[author] = author_id
                        self.graph.add_node(author_id, type='author', name=author)
                    self.graph.add_edge(author_nodes[author], paper_id, type='wrote')

        # Add keyword nodes and paper-keyword edges
        keyword_nodes = {}
        for idx, row in self.df.iterrows():
            paper_id = f"P{idx}"
            for keyword in row['Keywords_List']:
                if keyword and keyword != 'nan':
                    if keyword not in keyword_nodes:
                        keyword_id = f"K{len(keyword_nodes)}"
                        keyword_nodes[keyword] = keyword_id
                        self.graph.add_node(keyword_id, type='keyword', name=keyword)
                    self.graph.add_edge(paper_id, keyword_nodes[keyword], type='has_keyword')

        # Add citation edges (paper cites paper)
        # Using citation counts as a proxy for citation relationships
        # For more accurate citations, we would need the actual citation network data

        # Add co-authorship edges
        for idx, row in self.df.iterrows():
            authors = row['Authors_List']
            if len(authors) >= 2:
                for author1, author2 in combinations(authors, 2):
                    if author1 in author_nodes and author2 in author_nodes:
                        self.graph.add_edge(
                            author_nodes[author1],
                            author_nodes[author2],
                            type='co_author',
                            weight=1
                        )

        # Add journal nodes and paper-journal edges
        journal_nodes = {}
        for idx, row in self.df.iterrows():
            paper_id = f"P{idx}"
            journal = row['Source']
            if pd.notna(journal) and journal != 'nan':
                if journal not in journal_nodes:
                    journal_id = f"J{len(journal_nodes)}"
                    journal_nodes[journal] = journal_id
                    self.graph.add_node(journal_id, type='journal', name=journal)
                self.graph.add_edge(paper_id, journal_nodes[journal], type='published_in')

        # Add country nodes and paper-country edges
        country_nodes = {}
        for idx, row in self.df.iterrows():
            paper_id = f"P{idx}"
            country = row['Country']
            if country and country != 'Other':
                if country not in country_nodes:
                    country_id = f"C{len(country_nodes)}"
                    country_nodes[country] = country_id
                    self.graph.add_node(country_id, type='country', name=country)
                self.graph.add_edge(paper_id, country_nodes[country], type='from_country')

        print(f"Graph built with {self.graph.number_of_nodes()} nodes and {self.graph.number_of_edges()} edges")
        return self.graph

    def get_node_types(self):
        """Return count of each node type."""
        node_types = {}
        for node, data in self.graph.nodes(data=True):
            ntype = data.get('type', 'unknown')
            node_types[ntype] = node_types.get(ntype, 0) + 1
        return node_types


# ============================================================================
# PART 3: GRAPH ENRICHMENT AND ANALYSIS
# ============================================================================

class GraphAnalyzer:
    """Perform analysis on the knowledge graph."""

    def __init__(self, graph, df):
        self.graph = graph
        self.df = df

    def calculate_metrics(self):
        """Calculate various graph metrics."""
        metrics = {}

        # Basic metrics
        metrics['nodes'] = self.graph.number_of_nodes()
        metrics['edges'] = self.graph.number_of_edges()
        metrics['density'] = nx.density(self.graph)

        # Degree metrics
        degrees = dict(self.graph.degree())
        metrics['avg_degree'] = np.mean(list(degrees.values())) if degrees else 0
        metrics['max_degree'] = max(degrees.values()) if degrees else 0

        # Find top nodes by degree
        top_nodes = sorted(degrees.items(), key=lambda x: x[1], reverse=True)[:10]
        metrics['top_nodes'] = [(self.graph.nodes[n].get('name', n), d) for n, d in top_nodes]

        # Connected components
        components = list(nx.connected_components(self.graph.to_undirected()))
        metrics['num_components'] = len(components)
        metrics['largest_component_size'] = len(max(components, key=len)) if components else 0

        return metrics

    def find_top_authors(self, n=10):
        """Find top authors by number of publications."""
        author_pubs = {}
        for node, data in self.graph.nodes(data=True):
            if data.get('type') == 'author':
                # Count edges to paper nodes
                pubs = sum(1 for _, _, edge_data in self.graph.edges(node, data=True)
                          if edge_data.get('type') == 'wrote')
                author_pubs[data.get('name', node)] = pubs

        return sorted(author_pubs.items(), key=lambda x: x[1], reverse=True)[:n]

    def find_top_keywords(self, n=15):
        """Find most frequent keywords."""
        keyword_counts = {}
        for node, data in self.graph.nodes(data=True):
            if data.get('type') == 'keyword':
                keyword_counts[data.get('name', node)] = self.graph.degree(node)

        return sorted(keyword_counts.items(), key=lambda x: x[1], reverse=True)[:n]

    def find_top_countries(self, n=10):
        """Find countries with most publications."""
        country_counts = {}
        for node, data in self.graph.nodes(data=True):
            if data.get('type') == 'country':
                country_counts[data.get('name', node)] = self.graph.degree(node)

        return sorted(country_counts.items(), key=lambda x: x[1], reverse=True)[:n]

    def find_top_journals(self, n=10):
        """Find most productive journals."""
        journal_counts = {}
        for node, data in self.graph.nodes(data=True):
            if data.get('type') == 'journal':
                journal_counts[data.get('name', node)] = self.graph.degree(node)

        return sorted(journal_counts.items(), key=lambda x: x[1], reverse=True)[:n]

    def find_collaboration_network(self):
        """Find the largest collaboration component."""
        # Build author collaboration graph
        collab_graph = nx.Graph()

        for u, v, data in self.graph.edges(data=True):
            if data.get('type') == 'co_author':
                collab_graph.add_edge(u, v)

        # Find largest connected component
        if collab_graph.number_of_nodes() > 0:
            components = list(nx.connected_components(collab_graph))
            largest = max(components, key=len)
            return largest
        return set()

    def search_papers(self, query, search_type='title'):
        """Search for papers matching the query."""
        query = query.lower()
        results = []

        for idx, row in self.df.iterrows():
            paper_id = f"P{idx}"
            match = False

            if search_type == 'title' and query in str(row['Title']).lower():
                match = True
            elif search_type == 'author':
                for author in row['Authors_List']:
                    if query in author.lower():
                        match = True
                        break
            elif search_type == 'keyword':
                for keyword in row['Keywords_List']:
                    if query in keyword.lower():
                        match = True
                        break
            elif search_type == 'abstract' and query in str(row['Abstract']).lower():
                match = True
            else:  # all fields
                if (query in str(row['Title']).lower() or
                    query in str(row['Abstract']).lower() or
                    any(query in a.lower() for a in row['Authors_List']) or
                    any(query in k.lower() for k in row['Keywords_List'])):
                    match = True

            if match:
                results.append({
                    'paper_id': paper_id,
                    'title': row['Title'],
                    'authors': ', '.join(row['Authors_List'][:3]) + ('...' if len(row['Authors_List']) > 3 else ''),
                    'year': row['Year'],
                    'citations': row['Cites'],
                    'source': row['Source'],
                    'keywords': ', '.join(row['Keywords_List'][:5])
                })

        return results


# ============================================================================
# PART 4: VISUALIZATION MODULE
# ============================================================================

class GraphVisualizer:
    """Create visualizations of the knowledge graph."""

    def __init__(self, graph, df):
        self.graph = graph
        self.df = df

    def create_interactive_network(self, max_nodes=100, node_types=None):
        """Create an interactive network visualization using PyVis."""
        net = Network(height="750px", width="100%", bgcolor="#ffffff", font_color="black")

        # Filter nodes
        nodes_to_include = set()
        if node_types:
            for node, data in self.graph.nodes(data=True):
                if data.get('type') in node_types:
                    nodes_to_include.add(node)
        else:
            # Get highest degree nodes
            degrees = dict(self.graph.degree())
            nodes_to_include = set(sorted(degrees, key=degrees.get, reverse=True)[:max_nodes])

        # Add nodes
        for node in nodes_to_include:
            data = self.graph.nodes[node]
            node_type = data.get('type', 'unknown')

            # Node color mapping
            colors = {
                'paper': '#3498db',      # blue
                'author': '#e74c3c',     # red
                'keyword': '#2ecc71',    # green
                'journal': '#f39c12',     # orange
                'country': '#9b59b6',    # purple
                'unknown': '#95a5a6'     # gray
            }

            # Node size based on degree
            degree = self.graph.degree(node)
            size = 15 + min(degree * 2, 50)

            # Node title for tooltip
            title = f"{data.get('name', node)}\nType: {node_type}\nConnections: {degree}"
            if node_type == 'paper':
                title += f"\nCitations: {data.get('citations', 0)}\nYear: {data.get('year', 'N/A')}"

            net.add_node(
                node,
                label=data.get('name', node)[:20],
                title=title,
                color=colors.get(node_type, colors['unknown']),
                size=size,
                shape='dot'
            )

        # Add edges
        for u, v, data in self.graph.edges(data=True):
            if u in nodes_to_include and v in nodes_to_include:
                edge_type = data.get('type', 'related')
                edge_color = {
                    'wrote': '#e74c3c',
                    'has_keyword': '#2ecc71',
                    'published_in': '#f39c12',
                    'co_author': '#3498db',
                    'from_country': '#9b59b6'
                }.get(edge_type, '#95a5a6')

                net.add_edge(u, v, color=edge_color, title=edge_type)

        # Set physics layout
        net.set_options("""
        var options = {
            nodes: {
                font: { size: 12 }
            },
            edges: {
                smooth: { enabled: true }
            },
            physics: {
                stabilization: { iterations: 100 },
                barnesHut: { gravitationalConstant: -8000 }
            }
        }
        """)

        return net

    def create_publication_timeline(self):
        """Create a timeline of publications over years."""
        year_counts = self.df['Year'].value_counts().sort_index()

        fig = go.Figure(data=[
            go.Bar(
                x=year_counts.index.astype(int),
                y=year_counts.values,
                marker_color='#3498db',
                hovertemplate='Year: %{x}<br>Publications: %{y}<extra></extra>'
            )
        ])

        fig.update_layout(
            title='Publications Over Time',
            xaxis_title='Year',
            yaxis_title='Number of Publications',
            template='plotly_white',
            height=400
        )

        return fig

    def create_top_authors_chart(self, top_authors):
        """Create bar chart for top authors."""
        names, counts = zip(*top_authors) if top_authors else ([], [])

        fig = go.Figure(data=[
            go.Bar(
                x=list(names)[:10],
                y=list(counts)[:10],
                marker_color='#e74c3c',
                orientation='v',
                hovertemplate='Author: %{x}<br>Publications: %{y}<extra></extra>'
            )
        ])

        fig.update_layout(
            title='Top Authors by Publication Count',
            xaxis_title='Author',
            yaxis_title='Number of Publications',
            template='plotly_white',
            height=500,
            xaxis_tickangle=-45
        )

        return fig

    def create_top_keywords_chart(self, top_keywords):
        """Create horizontal bar chart for top keywords."""
        keywords, counts = zip(*top_keywords) if top_keywords else ([], [])

        fig = go.Figure(data=[
            go.Bar(
                x=list(counts)[:15],
                y=list(keywords)[:15],
                marker_color='#2ecc71',
                orientation='h',
                hovertemplate='Keyword: %{y}<br>Occurrences: %{x}<extra></extra>'
            )
        ])

        fig.update_layout(
            title='Top Keywords by Frequency',
            xaxis_title='Number of Occurrences',
            yaxis_title='Keyword',
            template='plotly_white',
            height=500
        )

        return fig

    def create_country_distribution(self, top_countries):
        """Create pie chart for country distribution."""
        countries, counts = zip(*top_countries) if top_countries else ([], [])

        fig = go.Figure(data=[
            go.Pie(
                labels=list(countries),
                values=list(counts),
                hole=0.3,
                marker=dict(colors=px.colors.qualitative.Set3),
                hovertemplate='%{label}<br>Publications: %{value}<br>Percentage: %{percent}<extra></extra>'
            )
        ])

        fig.update_layout(
            title='Publication Distribution by Country',
            template='plotly_white',
            height=450
        )

        return fig

    def create_citation_distribution(self):
        """Create histogram of citation distribution."""
        citations = self.df[self.df['Cites'] > 0]['Cites']

        fig = go.Figure(data=[
            go.Histogram(
                x=citations,
                nbinsx=30,
                marker_color='#f39c12',
                hovertemplate='Citations: %{x}<br>Count: %{y}<extra></extra>'
            )
        ])

        fig.update_layout(
            title='Citation Distribution',
            xaxis_title='Number of Citations',
            yaxis_title='Number of Papers',
            template='plotly_white',
            height=400
        )

        return fig

    def create_network_summary(self):
        """Create a summary dashboard of network metrics."""
        # Count node types
        node_types = {}
        for node, data in self.graph.nodes(data=True):
            ntype = data.get('type', 'unknown')
            node_types[ntype] = node_types.get(ntype, 0) + 1

        fig = go.Figure(data=[
            go.Bar(
                x=list(node_types.keys()),
                y=list(node_types.values()),
                marker_color=['#3498db', '#e74c3c', '#2ecc71', '#f39c12', '#9b59b6'],
                hovertemplate='Type: %{x}<br>Count: %{y}<extra></extra>'
            )
        ])

        fig.update_layout(
            title='Network Node Distribution by Type',
            xaxis_title='Node Type',
            yaxis_title='Number of Nodes',
            template='plotly_white',
            height=400
        )

        return fig


# ============================================================================
# PART 5: STREAMLIT APPLICATION
# ============================================================================

class KnowledgeGraphApp:
    """Main Streamlit application for Knowledge Graph Explorer."""

    def __init__(self):
        self.preprocessor = None
        self.graph = None
        self.analyzer = None
        self.visualizer = None
        self.df = None

    def run(self):
        st.set_page_config(page_title="Knowledge Graph Explorer", layout="wide", page_icon="📚")

        st.title("📚 Academic Knowledge Graph Explorer")
        st.markdown("---")

        # Sidebar
        with st.sidebar:
            st.header("📁 Data Upload")
            uploaded_file = st.file_uploader(
                "Upload Excel/CSV File",
                type=['xlsx', 'xls', 'csv'],
                help="Upload your academic publications dataset"
            )

            st.markdown("---")
            st.header("🔍 Search Options")
            search_type = st.selectbox(
                "Search Field",
                ['title', 'author', 'keyword', 'abstract', 'all fields']
            )
            search_query = st.text_input("Enter search query", placeholder="e.g., AI, blockchain, library")

            if st.button("🔎 Search", type="primary"):
                if search_query and self.analyzer:
                    self.display_search_results(search_query, search_type)

            st.markdown("---")
            st.header("🎨 Visualization Settings")
            max_nodes = st.slider("Max Nodes in Network", 50, 500, 150)
            node_types_filter = st.multiselect(
                "Show Node Types",
                ['paper', 'author', 'keyword', 'journal', 'country'],
                default=['paper', 'author', 'keyword']
            )

            if st.button("🔄 Refresh Graph", use_container_width=True):
                if self.graph:
                    st.session_state['refresh_graph'] = True

        # Main content area
        col1, col2 = st.columns([2, 1])

        with col1:
            st.subheader("📊 Knowledge Graph Overview")

            if uploaded_file is not None:
                if 'graph_built' not in st.session_state or st.session_state.get('refresh_graph'):
                    with st.spinner("Loading and processing data..."):
                        self.build_knowledge_graph(uploaded_file)
                    st.session_state['graph_built'] = True
                    st.session_state['refresh_graph'] = False

                if self.graph:
                    # Display statistics
                    stats = self.preprocessor.get_statistics() if self.preprocessor else {}

                    metric_cols = st.columns(4)
                    metrics_data = [
                        ("📄 Publications", stats.get('total_publications', 0)),
                        ("✏️ Authors", stats.get('unique_authors', 0)),
                        ("🏷️ Keywords", stats.get('unique_keywords', 0)),
                        ("📅 Years", stats.get('year_range', 'N/A'))
                    ]

                    for col, (label, value) in zip(metric_cols, metrics_data):
                        with col:
                            st.metric(label, value)

                    # Interactive network visualization
                    st.markdown("### 🌐 Interactive Knowledge Graph")
                    net = self.visualizer.create_interactive_network(
                        max_nodes=max_nodes,
                        node_types=node_types_filter if node_types_filter else None
                    )
                    net.save_graph("temp_network.html")
                    with open("temp_network.html", "r", encoding="utf-8") as f:
                        st.components.v1.html(f.read(), height=600)

                    # Analytics section
                    st.markdown("### 📈 Analytics Dashboard")
                    tab1, tab2, tab3, tab4 = st.tabs(
                        ["📅 Timeline", "🏆 Top Authors", "🔑 Keywords", "🌍 Countries"]
                    )

                    with tab1:
                        fig = self.visualizer.create_publication_timeline()
                        st.plotly_chart(fig, use_container_width=True)

                    with tab2:
                        top_authors = self.analyzer.find_top_authors(15)
                        fig = self.visualizer.create_top_authors_chart(top_authors)
                        st.plotly_chart(fig, use_container_width=True)
                        with st.expander("View full author list"):
                            for author, count in top_authors:
                                st.write(f"**{author}**: {count} publications")

                    with tab3:
                        top_keywords = self.analyzer.find_top_keywords(20)
                        fig = self.visualizer.create_top_keywords_chart(top_keywords)
                        st.plotly_chart(fig, use_container_width=True)

                    with tab4:
                        top_countries = self.analyzer.find_top_countries()
                        fig = self.visualizer.create_country_distribution(top_countries)
                        st.plotly_chart(fig, use_container_width=True)

            else:
                st.info("👈 Please upload your dataset file to begin")
                st.markdown("""
                ### Expected Data Format
                The dataset should contain columns such as:
                - **Title**: Paper title
                - **Authors**: Semicolon-separated author names
                - **Year**: Publication year
                - **Cites**: Number of citations
                - **Keywords**: Semicolon-separated keywords
                - **Source**: Journal/Conference name
                - **Publisher**: Publishing organization
                - **Abstract**: Paper abstract
                """)

        with col2:
            st.subheader("📋 Graph Statistics")

            if self.graph:
                metrics = self.analyzer.calculate_metrics()

                st.markdown("#### Network Metrics")
                st.metric("Total Nodes", metrics['nodes'])
                st.metric("Total Edges", metrics['edges'])
                st.metric("Network Density", f"{metrics['density']:.4f}")
                st.metric("Average Degree", f"{metrics['avg_degree']:.2f}")

                st.markdown("#### Node Type Distribution")
                node_types = self.visualizer.create_network_summary()
                st.plotly_chart(node_types, use_container_width=True)

                st.markdown("#### Top Connected Nodes")
                for node_name, degree in metrics['top_nodes'][:5]:
                    st.write(f"• {node_name}: {degree} connections")

                st.markdown("#### Recent Publications")
                recent_pubs = self.df.nlargest(5, 'Year') if self.df is not None else []
                for _, row in recent_pubs.iterrows():
                    with st.expander(f"📄 {row['Title'][:50]}..."):
                        st.write(f"**Authors**: {row['Authors'][:100]}...")
                        st.write(f"**Year**: {row['Year']}")
                        st.write(f"**Citations**: {row['Cites']}")
                        st.write(f"**Source**: {row['Source']}")
            else:
                st.info("Upload data to view statistics")

    def build_knowledge_graph(self, uploaded_file):
        """Build the knowledge graph from uploaded file."""
        self.preprocessor = DataPreprocessor()
        self.preprocessor.load_data(uploaded_file=uploaded_file)
        self.df = self.preprocessor.preprocess_data()

        builder = KnowledgeGraphBuilder(self.preprocessor.cleaned_df)
        self.graph = builder.build_graph()

        self.analyzer = GraphAnalyzer(self.graph, self.preprocessor.cleaned_df)
        self.visualizer = GraphVisualizer(self.graph, self.preprocessor.cleaned_df)

    def display_search_results(self, query, search_type):
        """Display search results."""
        st.markdown("---")
        st.subheader(f"🔍 Search Results: '{query}'")

        results = self.analyzer.search_papers(query, search_type)

        if results:
            st.success(f"Found {len(results)} matching papers")

            for i, result in enumerate(results[:20]):  # Limit to 20 results
                with st.expander(f"📄 {result['title']}"):
                    col1, col2 = st.columns([3, 1])
                    with col1:
                        st.write(f"**Authors**: {result['authors']}")
                        st.write(f"**Source**: {result['source']}")
                        st.write(f"**Keywords**: {result['keywords']}")
                    with col2:
                        st.write(f"**Year**: {result['year']}")
                        st.write(f"**Citations**: {result['citations']}")

            if len(results) > 20:
                st.info(f"Showing first 20 of {len(results)} results")
        else:
            st.warning("No results found. Try a different search query.")


# ============================================================================
# MAIN EXECUTION
# ============================================================================

def main():
    app = KnowledgeGraphApp()
    app.run()


if __name__ == "__main__":
    main()
```
